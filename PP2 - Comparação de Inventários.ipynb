```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# leitura dos DataFrames
df_24 = pd.read_csv('/content/Scan_Desenv_2024.csv',sep=',')
df_25 = pd.read_csv('/content/Scan_Desenv_2025.csv',sep=',')

# junção dos valores e separação dos valores não encontrados e novos
df_all = df_24.merge(df_25, on='Patrimonio', how='outer', indicator=True,suffixes=('_24','_25'))

dfu_24 = df_all[df_all['_merge'] == 'left_only']
dfu_24 = dfu_24.drop(columns=['_merge'])

dfu_25 = df_all[df_all['_merge'] == 'right_only']
dfu_25 = dfu_25.drop(columns=['_merge'])

# patrimonios que foram realocados
df_realoc = df_all[(df_all['_merge'] == 'both') & (df_all['Proprietario_24'] != df_all['Proprietario_25'])]
df_realoc = df_realoc.drop(columns=['_merge','Equipamento_25'])
df_realoc = df_realoc.rename(columns={'Equipamento_24':'Equipamento'})

# criando a visualização
plt.figure()
plt.bar(['Patrimonios Não Localizados','Patrimonios Novos','Patrimonios Realocados'],[len(dfu_24),len(dfu_25),len(df_realoc)],color=['#e34646','#58a7e8','#c176cf'])
plt.title('Comparação Inventário 2024 x Inventário 2025')
plt.ylabel('Quantidade de Patrimônios')
plt.tight_layout()
plt.show()
```
